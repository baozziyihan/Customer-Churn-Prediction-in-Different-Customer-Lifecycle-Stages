```python
teacher_detail['t_tag'] = teacher_detail['t_tag'].replace('', None)
teacher_detail['t_tag'] = teacher_detail['t_tag'].replace(np.nan, None)
teacher_detail
```

```python
# Convert the column values to strings 
teacher_detail['t_tag'] = teacher_detail['t_tag'].astype(str).replace('None', '')

# Split the strings on commas and flatten the lists
t_tag_list = teacher_detail['t_tag'].str.split(',').explode()

# Remove any empty strings from the list
t_tag_list = t_tag_list[t_tag_list != '']

# Convert the values to float
t_tag_list = t_tag_list.astype(float)

# Convert the list to a set to remove duplicates
t_tag_unique = set(t_tag_list)

# Convert the set back to a list
t_tag_unique_list = list(t_tag_unique)
t_tag_unique_list
```

```python
attribute_t_tag = attribute[attribute['attri_id'].isin(t_tag_unique_list)]
attribute_t_tag
```

```python
# Select important tags

# Create the 't_humorous' column
teacher_detail['t_humorous'] = teacher_detail['t_tag'].str.contains('113').astype(int)

# Create the 't_responsible' column
teacher_detail['t_responsible'] = teacher_detail['t_tag'].str.contains('114' or '341' or '339').astype(int)

# Create the 't_active' column
teacher_detail['t_active'] = teacher_detail['t_tag'].str.contains('263' or '347').astype(int)

# Create the 't_patient' column
teacher_detail['t_patient'] = teacher_detail['t_tag'].str.contains('345').astype(int)

# Create the 't_goodlook' column
teacher_detail['t_goodlook'] = teacher_detail['t_tag'].str.contains('264'or'346').astype(int)

# Create the 't_experienced' column
teacher_detail['t_experienced'] = teacher_detail['t_tag'].str.contains('350').astype(int)


teacher_detail = teacher_detail.drop('t_tag', axis=1)
teacher_detail
```
