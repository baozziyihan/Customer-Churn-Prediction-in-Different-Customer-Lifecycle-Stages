```python
import pandas as pd
```

```python
member_demographics = pd.read_csv('member_demographics.csv')
member_demand = pd.read_csv('member_demand.csv')
customer_reach = pd.read_csv('customer_reach.csv')
customer_conversion = pd.read_csv('customer_conversion.csv')
customer_repurchase = pd.read_csv('customer_repurchase.csv')
teacher_detail = pd.read_csv('teacher_detail.csv')
```

```python
# Customer_reach_final
member_detail = member_demographics.merge(member_demand, on='user_id',how='left')
customer_reach_final = member_detail.merge(customer_reach,on='user_id',how='left')
customer_reach_final.info()
```

```python
# Customer_conversion_final
customer_conversion = teacher_detail.merge(customer_conversion,on='teacher_id',how='right')
customer_conversion_final = member_detail.merge(customer_conversion, on='user_id',how='right')
customer_conversion_final.info()
```

```python
# Customer_repurchase_final
customer_repurchase = teacher_detail.merge(customer_repurchase,on='teacher_id',how='right')
customer_repurchase_final = member_detail.merge(customer_repurchase, on='user_id',how='right')
customer_repurchase_final.info()
```

```python
# Save the DataFrame to a CSV file
customer_reach_final.to_csv('customer_reach_final.csv', index=False)
customer_conversion_final.to_csv('customer_conversion_final.csv', index=False)
customer_repurchase_final.to_csv('customer_repurchase_final.csv', index=False)

print("saved to CSV file successfully.")
```
